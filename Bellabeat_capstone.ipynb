```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Loading the first dataset

```python
df1 = pd.read_csv('dailyActivity_merged.csv')
```

Checking the first five rows

```python
df1.head()
```

Checking the properties of the dataset

```python
df1.info()
```

Changing the ActivityDate column to datetime dtype and creating a column for day of the week

```python
df1['ActivityDate'] = pd.to_datetime(df1['ActivityDate'],format ='%m/%d/%Y')
df1['Day Of Week']=df1['ActivityDate'].dt.dayofweek
```

```python
dmap= {0:'Mon',1:'Tues',2:'Wed',3:'Thurs',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df1['Day Of Week'] = df1['Day Of Week'].map(dmap)
```

```python
df1.head()
```

Calculating the total amount of minutes spent and covering it to hours

```python
df1['TotalMinutes'] = df1['VeryActiveMinutes']+df1['FairlyActiveMinutes']+df1['LightlyActiveMinutes']+df1['SedentaryMinutes']
```

```python
df1['Hours'] = df1['TotalMinutes']/60
```

```python
df1.head()
```

Checking the number of unique IDs

```python
df1['Id'].nunique()
```

Checking the days in which the FITBIT app was used most

```python
#calculating total minutes
total_active = df1['VeryActiveMinutes'].sum()
total_fairly = df1['FairlyActiveMinutes'].sum()
total_lightly = df1['LightlyActiveMinutes'].sum()
total_sedentary = df1['SedentaryMinutes'].sum()

slices = [total_active,total_fairly,total_lightly,total_sedentary]
labels = ['Very Active Minutes','Fairly Active Minutes','Lightly Active Minutes','Sedentary Minutes']

plt.pie(slices,labels=labels,wedgeprops ={'edgecolor':'black'},autopct='%1.1f%%',explode=(0,0,0,0.1),shadow=True,colors=['gold','yellowgreen','lightcoral','lightskyblue'])
plt.tight_layout()
plt.title('Percentage of Activity in minutes')
```
